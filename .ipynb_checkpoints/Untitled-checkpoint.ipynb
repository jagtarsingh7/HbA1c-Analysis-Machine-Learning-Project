```python
"Encode into numbers"
def characters_to_numb(df, str_cols):
    for col in str_cols:
        df[col] = pd.Categorical(df[col])
        df[col] = df[col].cat.codes
    return df

obj_columns = list(data.loc[:, data.dtypes == object].columns)

processed_df = characters_to_numb(data, obj_columns)
```

```python
processed_df.head(20)
```

```python
cols = processed_df.columns

num_cols = processed_df._get_numeric_data().columns
num_cols
```

```python
list(set(cols) - set(num_cols))
```

```python
#dropped all the IDs columns
df = processed_df.drop('encounter_id', axis=1)
df = df.drop('admission_type_id', axis=1)
df = df.drop('discharge_disposition_id', axis=1)
df = df.drop('admission_source_id', axis=1)
df = df.drop('patient_nbr', axis=1)
```

```python
#check outliers and analysis
def find_outliers_IQR(df):

    q1=df.quantile(0.25)

    q3=df.quantile(0.75)

    IQR=q3-q1

    outliers = df[((df<(q1-1.5*IQR)) | (df>(q3+1.5*IQR)))]

    return outliers

outliers = find_outliers_IQR(df)
outliers.head(50)
```

```python
print("number of outliers: "+ str(len(outliers)))
print("max outlier value:")
print(str(outliers.max()))
```

```python
print("min outlier value:")
print(str(outliers.min()))
```

```python
#splitting the data.
X = df.drop('readmitted',axis=1)
y = df['readmitted']
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X)
X_scaled = scaler.transform(X)
```

```python
from sklearn.model_selection import train_test_split
```

```python
# split into train and test set
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, stratify=y, test_size=0.10, random_state=42)
```

```python
# create the classifier
classifier = RandomForestClassifier(n_estimators=100)

# Train the model using the training sets
classifier.fit(X_train, y_train)
```

```python
# predictin on the test set
y_pred = classifier.predict(X_test)
```

```python
# Calculate Model Accuracy
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

```python
#Our accuracy is around 57.5% which is bad. But we can always make it better.
```
